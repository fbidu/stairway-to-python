# Plantas e Árvores de Decisão


Uma árvore de decisão é uma ferramenta que ajuda a tomar decisões com base em várias opções e critérios. Pense nela como um mapa que você segue quando precisa escolher entre diferentes caminhos. Cada "nó" na árvore representa uma escolha a ser feita, e os "galhos" conectam essas escolhas de acordo com regras específicas. À medida que você percorre os galhos, você chega a uma "folha", que é a decisão final. As árvores de decisão são usadas para resolver problemas em várias áreas, como negócios, ciência e medicina, pois ajudam a tornar escolhas claras e compreensíveis, mesmo quando há muitos fatores envolvidos.

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# Carregar o conjunto de dados Iris
iris = load_iris(as_frame=True)
X = iris.data
y = iris.target
```

```python
X
```

```python
iris.target_names
```

```python
# Dividir o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Criar um classificador de árvore de decisão
clf = DecisionTreeClassifier()

# Treinar o classificador
clf.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = clf.predict(X_test)

# Calcular a acurácia das previsões
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia:", accuracy)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25, 20))
ax = fig.add_subplot(111)  # Crie um objeto Axes
plot_tree(clf, feature_names=iris.feature_names, class_names=list(iris.target_names), filled=True, ax=ax)
plt.show()
```
